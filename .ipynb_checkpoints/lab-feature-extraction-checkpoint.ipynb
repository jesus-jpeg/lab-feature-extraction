````python
customer_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
customer_df.head()
````

````python
cols = [col_name.lower().replace(' ', '_') for col_name in customer_df]
customer_df.columns = cols
customer_df.columns
````

````python
customer_df['effective_to_date'] = ["January" if x.startswith("1")==True else "February" for x in customer_df['effective_to_date']]
````

## 1. Selecting categoricals

````python
categoricals = customer_df.select_dtypes(np.object)
categoricals = categoricals.drop(['customer'], axis=1)
categoricals.head()
````

## 2. Plotting categoricals

````python
for col in categoricals:
    ax = sns.countplot(customer_df[col], order=customer_df[col].value_counts(ascending=True).index )
    plt.title(f'{col} count')
    plt.show()
````

## 4. Plotting Dates

We gonna do two cases, first one with number of clients by date, and second one with average claim amount by date.

````python
df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
df1 = df[["Effective To Date"]]
df2 = df[["Effective To Date","Total Claim Amount"]]
time = df1.groupby(df1.columns.tolist(),as_index=False).size()
tca = df2.groupby(df2["Effective To Date"])['Total Claim Amount'].mean().reset_index()
time.head()
````

````python
tca = pd.DataFrame(tca)
tca.head()
````

### Number of clients vs Date chart

````python
import datetime
import random
import matplotlib.dates as mdates
time["Effective To Date"] = pd.to_datetime(time["Effective To Date"], errors='coerce')
# sort DataFrame by date
time = time.sort_values(by=["Effective To Date"])

x_values = time["Effective To Date"]
y_values = time["size"]

# plot data
plt.plot_date(x_values, y_values, linestyle="solid", marker="")

# add labels and title
plt.xlabel('Date')
plt.ylabel('Size')
plt.title('Size vs. Date')

# rotate x-axis label
plt.xticks(rotation=45)

# show plot
plt.tight_layout()
plt.show()

````

**Another way of doing it** 

````python
# create timeseries plot
time.plot(x="Effective To Date", y="size")
plt.xlabel("Date",  size = 10)
plt.ylabel("Count", size = 10)
plt.title("Count of Effective to Date", size = 15)
````

### Avg Claim Amount vs Date

````python
import datetime
import random
import matplotlib.dates as mdates
tca["Effective To Date"] = pd.to_datetime(tca["Effective To Date"], errors='coerce')
# sort DataFrame by date
tca = tca.sort_values(by=["Effective To Date"])

x_values = tca["Effective To Date"]
y_values = tca["Total Claim Amount"]

# plot data
plt.plot_date(x_values, y_values, linestyle="solid", marker="")

# rotate x-axis label
plt.xticks(rotation=45)

plt.xlabel('Date')
plt.ylabel('Avg Claim Amount')
plt.title('Avg Claim Amount vs. Date')

# plot
plt.tight_layout()
plt.show()

plt.show()
````

### Merging both charts

````python
import pandas as pd
import matplotlib.pyplot as plt

# convert date column to datetime type
time["Effective To Date"] = pd.to_datetime(time["Effective To Date"], errors='coerce')
tca["Effective To Date"] = pd.to_datetime(tca["Effective To Date"], errors='coerce')

# sort DataFrame by date
time = time.sort_values(by=["Effective To Date"])
tca = tca.sort_values(by=["Effective To Date"])

# create figure and axes
fig, ax = plt.subplots()

# extract x and y values for chart 1
x_values_1 = time["Effective To Date"]
y_values_1 = time["size"]

# extract x and y values for chart 2
x_values_2 = tca["Effective To Date"]
y_values_2 = tca["Total Claim Amount"]


# plot chart 1
ax.plot_date(x_values_1, y_values_1, linestyle="solid", marker="", label="Quantity of customers")

# plot chart 2
ax.plot_date(x_values_2, y_values_2, linestyle="solid", marker="", label="Avg Claim Amount")

# add labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Measurement (units)')
ax.set_title('Size and Avg Claim Amount vs. Date')

# rotate x-axis label
plt.xticks(rotation=45)

# add legend
ax.legend()

# show plot
plt.tight_layout()
plt.show()
````
